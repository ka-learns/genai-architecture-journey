# Visualizing Tokenization in Hugging Face Transformers
Understanding how models see text.

```python
from transformers import AutoTokenizer
import pandas as pd
```

Each model has its own tokenizer, GPT2 uses BPE

```python
tokenizer = AutoTokenizer.from_pretrained("gpt2")
```

Tokenize a sentence

```python
text = "What is this"
tokens = tokenizer.tokenize(text)
token_ids = tokenizer.convert_tokens_to_ids(tokens)
```

Visualize the tokens from BPE

```python
pd.DataFrame({
    "Token": tokens,
    "Token_ID": token_ids
})
```

Lets try with T5 which uses SentencePiece

```python
tokenizer = AutoTokenizer.from_pretrained("t5-small")
tokenizer.tokenize(text)
tokens = tokenizer.tokenize(text)
token_ids = tokenizer.convert_tokens_to_ids(tokens)
```

Lets visualize T5

```python
pd.DataFrame({
    "Token": tokens,
    "Token_ID": token_ids
})
```
